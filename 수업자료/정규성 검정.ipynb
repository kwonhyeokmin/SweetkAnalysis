# 1. 정규성 검정(Normality Test)

> 데이터셋의 분포가 정규성분포를 따르는지 검사하는 것으로, 여러 통계적 방법론이 정규분포를 가정하고 있기때문에 데이터를 분석하기전에 필수적으로 요구되는 검정

### 정규성 검정의 종류

1. Shapiro-Wilks Test
    * 표본수가 2000미만인 데이터셋에 적합한 정규성 검정방법
2. Kolmogorove-Smimov Test
    * 표본수가 2000초과인 데이터셋에 적합한 정규성 검정방법
3. Quantile-Quantile plot
    * 시각적 검정법

```python
# 필요 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import seed
%matplotlib inline
```

```python
from pandas import Series

# 데이터 생성
seed(1) # seed값 고정
N = 2000
data = Series(np.random.randn(N))
```

```python
# 데이터 시각화
plt.hist(data, bins=100, density=True, color='blue')
data.plot.kde(color='red')
plt.show()
```

### Shapiro-Wilk Test

```python
from scipy.stats import shapiro

stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

### Kolmogorov-Smirnov test

```python
from scipy.stats import kstest

stats, p = kstest(data, 'norm')
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

### Quantile-Quantile plot

```python
from scipy.stats import probplot

plt.figure(figsize=(7,7))
probplot(data, plot=plt)
plt.axis("equal")
plt.show()
```
