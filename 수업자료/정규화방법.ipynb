# 2. 정규화 방법

```python
# 필요라이브러리 임포트
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 데이터 생성
df = pd.DataFrame({
    'x1': np.concatenate([np.random.normal(20, 1, 1000), np.random.normal(1, 1, 25)]),
    'x2': np.concatenate([np.random.normal(30, 1, 1000), np.random.normal(50, 1, 25)]),
})

df.plot.kde()
plt.show()
```

### 최대 최소 정규화(Min-Max Scaler)

> 분포의 최대값과 최소값을 이용하며 가장 널리쓰이는 정규화방법입니다.

$$MinMaxScaler(x_i) = \dfrac{x_i – min(x)}{max(x) – min(x)}$$

- 0에서 1의 범위를 가집니다(또는 -1에서 1사이의 값을 가질수도 있습니다).
- 분포가 가우시안 분포가 아니거나 표준편차가 작은 등, 표준정규화가 잘 되지 않는 경우에 잘 작동합니다.
- 이상치에 민감하기 때문에 이상치가 많은경우 로버스트 정규화를 활용하는것이 좋습니다.

```python
from sklearn.preprocessing import MinMaxScaler

minmaxscaler = MinMaxScaler()
data_tf = minmaxscaler.fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1', 'x2'])
df.plot.kde()
plt.show()
```

### 로버스트 정규화( Robust Scaler)

$RobustScaler(x_i) = \dfrac{x_i – Q_1(x)}{Q_3(x) – Q_1(x)}$

* 최대최소 정규화와 비슷하지만 최대값과 최소값이 아닌 사분편차 값을 활용합니다
* 최대최소 정규화에 비해 이상치에 덜 민감합니다

```python
from sklearn.preprocessing import RobustScaler

robustscaler = RobustScaler()
data_tf = robustscaler.fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1','x2'])
df.plot.kde()
plt.show()
```
